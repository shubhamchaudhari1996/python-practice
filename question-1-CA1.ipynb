```python
# Importing necessary libraries

import re
import datetime
```

```python
class Employee:

    # Assigning regex expression to a variable
    regexInt = '^[0-9]*$'
    regexChar = r"[\sa-zA-Z]+$"
    regexCombine = "^[0-9A-Za-z]*$"

    # Validation for user input as a number
    def checkNumber(msg):
        while True:
            try:
                myValue = float(input(msg))
            except ValueError:
                print("Invalid Input, Please re-enter")
                continue
            else:
                return myValue

    # Validations on user input for employee name
    while True:
        employeeName = input('Employee Name: ')
        if not re.match(regexChar,employeeName) or re.match(regexInt,employeeName) or employeeName.strip()=='':
            print('Please enter a valid employee name')
            continue
        else:
            break

    # Validations on user input for employee number
    while True:
        employeeNumber = input('Employee Number: ')
        if not re.match(regexCombine,employeeNumber):
            print('Please enter a valid employee number')
            continue
        else:
            break

    # validation on user input for week ending
    while True:
        week = input('Week ending: ')
        try:
            datetime.datetime.strptime(week,"%d/%m/%Y")
            break
        except ValueError:
            print('Invalid Input, please use dd/mm/yyyy format')
            continue

    # Getting hours worked from user
    hoursWorked = checkNumber("Number of hours worked: ")

    # Getting rate per hour from user
    hourlyRate = checkNumber('Hourly Rate: ')

    # Validations on user input for over time rate
    if hoursWorked > 37.5:
        overTime = checkNumber('OverTime Rate: ')
    else:
        overTime = 0

    # Validations on user input for normal tax rate
    while True:
        taxRate = checkNumber('Standard Tax Rate: ')
        if taxRate >= 100:
            print('Invalid Input, Please re-enter')
            continue
        else:
            break

    # Validations on user input for overtime tax rate
    while True:
        overTax = 0
        if hoursWorked > 37.5:
            overTax = checkNumber('OverTime Tax Rate: ')
            if overTax >= 100:
                print('Invalid Input, Please re-enter')
                continue
            else:
                break
        break

    # Method to calculate earning for standard working hours
    def calculateEarning(self):
        self.earning = 0
        if self.hoursWorked >= 37.5:
            self.earning = 37.5 * self.hourlyRate
            return self.earning
        else:
            self.earning = self.hoursWorked * self.hourlyRate
            return self.earning

    # Method to calculate overtime earning
    def calculateOverTime(self):
        self.extraTime = 0
        self.overEarn = 0
        if self.hoursWorked > 37.5:
            self.extraTime = self.hoursWorked - 37.5
            self.hoursWorked = 37.5
            self.overEarn = self.extraTime * self.overTime
            return self.overEarn
        else:
            return 0

    # Method to calculate tax on standard earning
    def calculateTaxDeduction(self):
        self.tax = (self.earning * self.taxRate)/100
        return self.tax

    # Method to calculate tax on overtime earning
    def calculateOverTimeTax(self):
        self.otax = 0
        if self.overTime > 0:
            self.otax =  (self.overEarn * self.overTax)/100
            return self.otax
        else:
            return 0

    # Method to calculate total earnings
    def totalEarning(self):
        self.total = 0
        if self.hoursWorked > 37.5:
            self.total = self.earning + self.overEarn
            return self.total
        else:
            self.total = self.earning
            return self.total

    # Method to calculate total tax on earnings
    def totalTax(self):
        self.taxTotal = 0
        if self.hoursWorked >= 37.5:
            self.taxTotal = self.tax + self.otax
            return self.taxTotal
        else:
            self.taxTotal = self.tax
            return self.taxTotal

    # Method to calculate net payment after tax deduction
    def netPayment(self):
        self.netpay = 0
        if self.hoursWorked >= 37.5:
            self.netpay = self.total - self.taxTotal
            return self.netpay
        else:
            self.netpay = self.earning - self.tax
            return self.netpay

    # Method to print payslip
    def showPayslip(self):
        print("\n\t\t\t\t\t\t\t\tPAYSLIP\n");
        print('WEEK ENDING ',self.week)
        print('Employee: ',self.employeeName)
        print('Employee Number: ',self.employeeNumber)
        print('\t\t\t\t\t\tEarnings\t\t\t\t\t\t\t\tDeduction')
        print('\t\t\t\t\t\tHours\t\t\tRate\t\t\tTotal')
        print('Hours (normal)\t\t\t',self.hoursWorked,'\t\t\t',self.hourlyRate,'\t\t',self.earning,'Tax @',self.taxRate,'%',self.tax)
        print('Hours (overtime)\t\t', self.extraTime, '\t\t\t', self.overTime, '\t\t', self.overEarn, 'Tax @',self.overTax,'%', self.otax)
        print('\n\t\t\t\tTotal Pay:\t\t\t\t\t\t\t\t\t',self.total)
        print('\t\t\t\tTotal deduction:\t\t\t\t\t\t\t',self.taxTotal)
        print('\t\t\t\tNet pay:\t\t\t\t\t\t\t\t\t',self.netpay)

# object initialization
e = Employee()

# Calling object
e.calculateEarning()
e.calculateOverTime()
e.calculateTaxDeduction()
e.calculateOverTimeTax()
e.totalEarning()
e.totalTax()
e.netPayment()
e.showPayslip()
```

```python
# Enter Payslip Details

emp_name = str(input("Enter employee name: "))
emp_number = str(input("Enter employee number: "))
week_end = str(input("Enter week ending (DD/MM/YYYY): "))
hours_worked = float(input("Number of hours worked: "))
overtime_hours_worked = float(input("Number of overtime hours worked: "))
hourly_rate = float(input("Enter hourly rate: "))
standard_taxrate = int(input("Enter Standard Tax Rate: "))
overtime_taxrate = int(input("Enter Overtime Tax Rate: "))
```

```python
# afafa
overtime_rate = hourly_rate * 1.5

# azfaaf
total_pay = hours_worked * hourly_rate

#zdfvsdgs
tax_20 = round((total_pay/100)*standard_taxrate,2)

# zdfdadfa
overtime_totalpay = overtime_hours_worked * overtime_rate

# dsfsxgg
tax_50 = round((overtime_totalpay/100)*overtime_taxrate,2)

# sdfsfsf
total = round(total_pay + overtime_totalpay, 2)

# dgfsdsgg
total_deduction = round(tax_20 + tax_50,3)

# sdgsgsg
net_pay = round(total - total_deduction,3)
```

```python
# Printing final output

print("PAYSLIP".center(100))
print("WEEK ENDING", week_end)
print("Employee:",emp_name)
print("Employee Number:",emp_number)
print("Earnings".center(52),"Deductions")
print("Hours       Rate Total  ".center(68))
print("Hours (normal)",'{:^18}'.format(str(hours_worked)),hourly_rate,round(total_pay,2),"       Tax @ 20% ",tax_20)
print("Hours (overtime)",'{:^16}'.format(str(overtime_hours_worked)),round(overtime_rate,2),round(overtime_totalpay,2),"       Tax @ 50% ",tax_50)
print("\n")
print("Total pay:".center(52),str(total).center(29))
print("Total deductions: ".center(60),str(total_deduction).center(15))
print("Total pay:".center(52),str(net_pay).center(32))
```
